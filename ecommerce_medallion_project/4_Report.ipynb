```python
df = spark.table("ecommerce.gold.fact_transactions_denorm")
display(df)
```

Ecommerce sales dashboard

```python
sales_by_date = df.groupBy("transaction_date").agg({"net_amount": "sum"}).withColumnRenamed("sum(net_amount)", "total_sales")
display(sales_by_date)
```

```python
customers_per_category = df.groupBy("category_name").agg({"customer_id": "count"}).withColumnRenamed("count(customer_id)", "total_customers")
display(customers_per_category)
```

```python
orders_per_channel = df.groupBy("channel").agg({"transaction_id": "count"}).withColumnRenamed("count(transaction_id)", "total_orders")
display(orders_per_channel)
```

```python
net_amount_by_country_region = df.groupBy("customer_country", "customer_region").agg({"net_amount": "sum"}).withColumnRenamed("sum(net_amount)", "total_net_amount")
display(net_amount_by_country_region)
```

```python
from pyspark.sql.functions import date_trunc

monthly_sales_trend = df.groupBy(date_trunc("month", "transaction_date").alias("month")).agg({"net_amount": "sum"}).withColumnRenamed("sum(net_amount)", "total_sales")
display(monthly_sales_trend)
```

```python
sales_by_coupon_discount = df.groupBy("coupon_code", "discount_percent").agg({"net_amount": "sum"}).withColumnRenamed("sum(net_amount)", "total_sales")
display(sales_by_coupon_discount)
```

```python
from pyspark.sql.functions import dayofweek

sales_by_day_of_week = df.groupBy(dayofweek("transaction_date").alias("day_of_week")).agg({"net_amount": "sum"}).withColumnRenamed("sum(net_amount)", "total_sales")
display(sales_by_day_of_week)
```
